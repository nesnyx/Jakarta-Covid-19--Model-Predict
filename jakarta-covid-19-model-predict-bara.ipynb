```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../input/indonesia-coronavirus-cases/jakarta.csv')
df
```

```python
X = pd.get_dummies(df.drop(['pdp_process', 'pdp_end', 'pdp_total'], axis=1))
y = df['pdp_total'].apply(lambda x: 1 if x=='Yes' else 0 )
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size =.1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=6, activation="relu",input_dim=len(X_train.columns)))
model.add(Dense(units=3, activation="softmax"))
model.add(Dense(units=1, activation="sigmoid"))
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics="accuracy")
```

```python
model.fit(X_train,y_train, epochs=20, batch_size = 2)
```

```python
y_hat = model.predict(X_test)
y_hat =  [0 if val <0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```
